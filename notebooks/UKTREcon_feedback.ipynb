# Feedback and reflections on UK TREcon 2023 

_This page is based on the feedback from a total of 25 respondents to survey sent out following the event, as well as some reflections on the event from the Community Management working group of the UK TRE community._

## Overall feedback on the day
Feedback from participants on the day as a whole was overwhelmingly positive, with participants noting that the at-capacity event was engaging, had a real buzz and was an excellent opportunity to share thoughts and ideas with large number of people from across the UK TRE landscape. A number of participants commented on the overall usefulness and relevance of the material presented on the day and thought the day was generally well-organised. With almost 100 in-person attendees and another 100 or so online, the event met its objective of being the largest gathering of people involved in TREs in the UK, and generally had good representation across the landscape (although one participant noted that representation was mostly from academia, and lacked NHS Secure Data Environment participation). The community management group reflected that the day was focussed mostly on the technical side of building and maintaining TREs, and that more could be done in future to bring in more researchers and greater public involvement and engagement. How the content of the day would need to be tweaked to make it accessible to less technically-minded folk would need to be thought carefully about.

```python
# Read in the MC feedback data
data = pd.read_csv('../data/UKTREcon_Feedback.csv', skiprows = [1])
```

## Accessibility
Online attendees said that support for online attendance during the morning session was very good, although a number remarked that having Zoom links sent out more in advance would have been beneficial with some participants also favouring MS Teams for these kinds of events because they can be more easily linked to an Eventbrite page to automate meeting link sharing.

A  number of in-person participants noted issues at the venue around ventilation and room temperature on what was an exceptionally hot day. There were some accessibility issues for in-person attendees with disabilities because of a broken lift. In terms of location, some attendees noted that Swansea was one of the more difficult cities for many in the UK to reach, and one participant favoured the city centre campus over the Bay campus. However, maintaining the connection to UK RSEcon had logistical benefits in organising and running the event, and encouraged attendance and participation of the RSE community out of which the UK TRE community came.

## Feedback on the keynote presentation
The Keynote presentation was commended as 'excellent', and a clear description of HDR UK's activities, the wider context, and interconnectivity. A few participants felt that the talk could have done more to explicitly address the question in the title on the technology ecosystem helping TREs to work together. The consensus was that the answer to this question is 'Yes' - a key challenge for the UK TRE community going forward will be to come up with ideas on _how_ the technical ecosystem can be convened to address core challenges on areas such as identity management.

## Feedback on the lightning talks
The diversity of the lightning talks was seen as a key strength, and started to build a more complete picture of the ecosystem beyond just the research infrastructure itself to touch on important wider topics including training and capacity building, researcher's perspectives on TRE utility, as well as the research that they can enable. Generally the feedback on the content was very good. Providing time constraints for the talks, but giving speakers the flexibility to choose how they spent that time (including the number of slides to use), worked well. 
One participant noted, given the time constraints, that talks mostly focused on 'here we are' and 'this is what we have' rather than what is challenging or what is needed to go forward. 

The Community Management group was grateful to all the speakers for keeping to time so conscientiously and without compromising on the quality of the talks - it was clear that a lot of effort had been made by the speakers to do so and it paid off for their audiences. Online attendees in particular expressed that it was challenging, given the very high turnover rate of talks, to keep track of who was who, and that the absence of any questions gave the talks a bit of a 'broadcast' feel. The quick-fire format allowed all those that wanted to give a lightning presentation the opportunity to do so, although this necessitated splitting the rooms into two parallel sessions. Recording both the sessions was important to mitigate the impact of this for those that were keen on content from the parallel session that they could not attend, however some participants felt it would have been beneficial to have given more time for the lightning talk sessions and not to run them in parallel.

## Feedback on breakout discussions
The consensus from participants was that the discussions were very interesting, served as useful starting points particularly around different ways TREs handle different types of problems, and provided an excellent opportunity for networking and a deeper dive into specific issues. Crowdsourcing the discussion topics from attendees was commended, but one participant felt like the process for choosing the topics was a little confusing and disjointed, and in instances when the person posing the topic question was not at the breakout table participants sometimes struggled to understand the intent or meaning of the issue. Some participants felt that it was unclear how to contribute to the roadmap, and providing more instructions or a template up front about how conversations could be structured (e.g. defining the 'problem', 'potential solutions', and 'ways to collaborate') could have helped focus discussions. Skipping the 'readout session' that typically follows breakout discussions was welcomed by one participant, however in the absence of a read out summary it was felt that each breakout discussion should produce at a written summary that captured the key messages from the discussion.

## Some reflections about what could be done differently next time
**Accessibility**: 
- The organisers should consider including information to attendees about what lift access is available at the venue, and to send updates if lifts are broken and what alternatives there are. Wherever possible, breakout rooms and lunch rooms should be on the same floor and as near as possible to the main session room.
- For any online sessions, the organisers should consider switching to MS Teams, which integrates more easily with Eventbrite for automated link sharing and other features (including permission to use from organisational policies for individuals working for the civil service and other relevant public bodies such as the Office for National Statistics). 
- The location for the next event should be chosen carefully to allow as many people to attend in person as possible - many attendees had to spend an overnight in Swansea for the one-day event, at a more accessible location attendees could attend a two-day session for a similar overall investment in time.

**Keynote**:
- The organisers should consider whether to include more than one keynote presentation, especially if the event is include wider representation from stakeholders such as researchers who use (rather than build and manage) research infrastructure, and members of the public.

**Lightning talks**: 
- A small amount of additional time could be allocated for (at least) one question following each talk, or else some time for discussion at the end of a section, for clarification or else to seed a bigger discussion later on as well as to give the audience a minute or two to 'digest' what they heard and gather their composure for the next quick-fire and typically information-dense talk. 
- Grouping the talks together thematically might also help the audience to assimilate the information more easily. If there's enough time, the organisers should consider bringing all the talks together into a single session (with breaks).

**Breakout discussions**:
- The person posing the topic question for a breakout discussion should be asked to sit at the table where their topic is being discussed to help guide the discussion and clarify the meaning of the question
- Provide more instructions or a template up front about how conversations could be structured (e.g. defining the 'problem', 'potential solutions', and 'ways to collaborate') to help focus discussions.
- Ask groups to provide a written summary of what was discussed (as well as the longer form notes), especially if there is no 'readout' session

**Networking opportunities**:
- Consider making the event longer (e.g. two days) to help give more time for informal networking opportunities. Holding the event in centrally located city in the UK with good transport links should help reduce travel time for participants.
